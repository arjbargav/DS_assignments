```python
observed = [
    [271, 29],
    [267, 33],
    [269, 31],
    [280, 20]
]
observed
```

```python
import scipy.stats as sc
```

```python
alpha = 0.05

chi2,p,dof,expected = sc.chi2_contingency(observed)
```

```python
p, alpha
```

## p-value > alpha, hence we fail to reject the null hypothesis.

## Therefore, the % of defective customer order forms are same in all the 4 centers
