## Null Hypothesis (H0): Mean Diameter of Unit A = Mean Diameter of Unit B
## Alternate Hypothess (Ha): Mean Diameter of Unit A != Mean Diameter of Unit B

## 1. Use two sample t-test
## 2. Two-tail test
## 3. Alpha = 0.05 (significance level)
## 4. Confidence interval: 0.025 - 0.975

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sc
%matplotlib inline
```

```python
df = pd.read_csv('Cutlets.csv')
df
```

```python
df.describe()
```

```python
alpha = 0.05
CI = [0.025,0.975]

t_stat, p_value = sc.ttest_ind(df['Unit A'],df['Unit B'])
p_value, CI, t_stat
```

## We see that, 
## p-value > alpha, reject Ha or fail to reject H0
## Hence, mean of dia of Unit A is statistically equal to that of Unit B
