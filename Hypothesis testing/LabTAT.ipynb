```python
import pandas as pd
import scipy.stats as sc
```

```python
df = pd.read_csv('LabTAT.csv')
```

```python
df
```

## Null hypothesis (H0): There is no difference in the collected sample sets
## Alternate hypothesis (Ha): There is a significant difference between the 4 sample sets
## Significance level, alpha = 0.05
## Test: ANOVA-one way

```python
alpha = 0.05

f_stat, p_value = sc.f_oneway(df['Laboratory 1'],df['Laboratory 2'],df['Laboratory 3'],df['Laboratory 4'])
p_value, alpha, f_stat
```

## p-value < alpha
## Therefore, we reject H0, and conclude that there is a significant difference between the TAT among the 4 sample sets.
