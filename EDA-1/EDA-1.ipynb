## Outlier detection

```python
sns.boxplot(df)
```

```python
df.head(1)
```

**We see that many columns have outliers**

```python
#Create a function that returns the upper and lower extremes in an IQR-plot

def find_outliers(data, colname):
    q1 = data[colname].quantile(0.25)
    q3 = data[colname].quantile(0.75)
    iqr = q3 - q1
    
    lower = q1 - (1.5 * iqr)
    upper = q3 + (1.5 * iqr)
    return lower, upper
```

### Outlier capping (replacing all outliers with extreme values)

```python
for col in df.columns:
    low_ext, up_ext = find_outliers(df, col)            # Find upper and lower extremes of each column
    try:
        df.loc[df[col] > up_ext, col] = up_ext          # Replace the outlier with the extreme values respectively
        df.loc[df[col] < low_ext, col] = low_ext
    except:
        pass
```
