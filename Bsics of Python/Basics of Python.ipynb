```python
# input a number
x = int(input("Enter a number\n"))

#Divisor variable
divisor = 1

# For numbers starting from 1 to x, we check if division yeilds remainder zero
for i in range(2,x):
    if(x%i==0):
        divisor = i

if(divisor!=1):
    print(f'{x} is not a prime number')
else:
    print(f'{x} is a prime number')
        
```

```python
#Use random module

import random as rm

a = rm.randrange(1,1000)
b = rm.randrange(1,1000)

prod = int(input(f'Enter the product of {a} and {b}\n'))

if(prod==(a*b)):
    print("Correct answer!")
else:
    print("Wrong answer")
```

```python
# Create a list of numbers from 100 to 200
numbers = list(x for x in range(100,201))

# We are printing the squares of all even numbers
for i in numbers:
    if(i%2==0):
        print(i*i)
```

```python
#Enter a sentence
text = 'This is a sample text. This text will be used to demonstrate the word counter.'

#Use split() to create a list of words
words = text.split()

#Empty dictionary to add words and counts
wcount = {}

for word in words:
    word = word.strip('.,?')
    wcount[word] = wcount.get(word, 0) + 1

for word, count in wcount.items():
    print(f'{word}: {count}')
```

```python
def is_palindrome(text):
    text = text.lower()

    if(text==text[::-1]):
        print(f'{text} is a palindrome')
    else:
        print(f'{text} is not a palindrome')

text = input("Enter a sentence:\n")

is_palindrome(text)
```
