## Feature Scaling for numerical columns

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

Use MinMaxScaler for the below columns as they have a nearly normal distribution - 
- age
- fnlwgt
- education_num
- hours_per_week
- capital_gain
- capital_loss

Use OneHotEncoder for those columns with less than 5 categories
- sex

Use LabelEncoder for the columns having more than 5 categories
- workclass
- education
- marital_status
- occupation
- relationship
- race
- native_country
- income (target column, therefore using label encoder to maintain the values in one column itself)


```python
df.head(1)
```

```python
# Normalization - This method is used as the chi2 test requires all values to be non-negative.
mn = MinMaxScaler()

df[['capital_gain','capital_loss','age','fnlwgt','education_num','hours_per_week']] = mn.fit_transform(df[['capital_gain','capital_loss','age','fnlwgt','education_num','hours_per_week']])
df.head()
```

```python
# One-Hot encoding
ohe = OneHotEncoder()

ohe_data_sex = ohe.fit_transform(df[['sex']])
trans_df = pd.DataFrame(ohe_data_sex.toarray(), columns=ohe.get_feature_names_out(['sex']))

trans_df.shape
```

```python
df.drop(columns=['sex'], inplace=True)
```

```python
df = pd.concat([df, trans_df], join='inner' ,axis=1)
df
```

```python
# Move the 'income' column to the last
income_col = df.pop('income')
df['income'] = income_col
```

```python
# Label encoding

le = LabelEncoder()

for i in ['workclass','education','marital_status','occupation','relationship','race','native_country','income']:
    df[i] = le.fit_transform(df[i])

df.head()
```

### **Pros and Cons of Label Encoding**
Pros:
- Simple and easy to implement for categorical columns where the order of category has importance.
- No change in dimensionality.
Cons:
- Bias, algorithms may consider higher categorical values with higher importance.

### **Pros and Cons of One-Hot Encoding**
Pros:
- Useful in situation where order of the categories is not important.
- Easy to interpret.
Cons:
- Increased dimenionality.
- Preferable for Nominal columns.

## **Isolation Forest for outlier detection**

```python
from sklearn.ensemble import IsolationForest
```

```python
df.head(3)
```

```python
df['fnlwgt'].max(), df['fnlwgt'].min()
```

```python
# As the 'fnlwgt' column has a -inf value, we need to replace this with the maximum value of float32 inorder to perform Isolation Forest

min_float = np.finfo(np.float32).min
min_float
```

```python
# Replace -inf with max_float

df['fnlwgt'].replace(-np.inf, min_float, inplace=True)
```

```python
iso = IsolationForest()

iso.fit(df)
```

```python
# Predict the outliers

outliers = iso.predict(df)

outliers
```

```python
df['outliers'] = outliers
df
```

```python
# Filter out all the outliers

df[df['outliers'] == -1]
```

```python
# Count of outliers

df['outliers'].value_counts()
```

We have 6073 data points as outliers as found by Isolation Forest algorithm

How outliers affect models performance:
1) Outliers lead to severe underfitting of the model leading to poor predictions.
2) Training model residuals/errors will be substantially high in the presence of outliers.
3) Features will be mislead into appearing more or less important during analysis
